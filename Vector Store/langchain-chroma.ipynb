```python
from langchain_openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma
```

```python
from langchain.schema import Document

doc1 = Document(
    page_content="""
Inception is a science fiction thriller directed by Christopher Nolan.
The film explores the concept of entering people's dreams to steal or
implant ideas. Dom Cobb, a skilled thief, is given a chance to erase his
criminal past by performing inception instead of extraction.

The movie is known for its complex narrative, visual effects, and themes
of reality, time, and memory.
""",
    metadata={
        "title": "Inception",
        "genre": "Science Fiction",
        "director": "Christopher Nolan",
        "release_year": 2010,
        "rating": 8.8,
        "language": "English"
    }
)

doc2 = Document(
    page_content="""
The Dark Knight is a superhero film featuring Batman as he faces the Joker,
a criminal mastermind who seeks to create chaos in Gotham City. The movie
focuses on moral dilemmas, justice, and the thin line between hero and villain.

Heath Ledger’s portrayal of the Joker is widely regarded as iconic.
""",
    metadata={
        "title": "The Dark Knight",
        "genre": "Action",
        "director": "Christopher Nolan",
        "release_year": 2008,
        "rating": 9.0,
        "language": "English"
    }
)

doc3 = Document(
    page_content="""
Interstellar is a science fiction drama that follows a group of astronauts
traveling through a wormhole in search of a new home for humanity. The film
explores themes of love, sacrifice, time dilation, and survival.

It combines scientific concepts with emotional storytelling.
""",
    metadata={
        "title": "Interstellar",
        "genre": "Science Fiction",
        "director": "Christopher Nolan",
        "release_year": 2014,
        "rating": 8.6,
        "language": "English"
    }
)

doc4 = Document(
    page_content="""
Titanic is a romantic drama based on the real-life sinking of the RMS Titanic.
The story follows Jack and Rose, two passengers from different social classes
who fall in love during the ship’s maiden voyage.

The film is known for its emotional depth and historical elements.
""",
    metadata={
        "title": "Titanic",
        "genre": "Romance",
        "director": "James Cameron",
        "release_year": 1997,
        "rating": 7.9,
        "language": "English"
    }
)

doc5 = Document(
    page_content="""
The Matrix is a science fiction action film that depicts a dystopian future
where humanity is unknowingly trapped inside a simulated reality. Neo, a
computer hacker, discovers the truth and joins a rebellion against machines.

The movie is famous for its philosophical themes and innovative action scenes.
""",
    metadata={
        "title": "The Matrix",
        "genre": "Science Fiction",
        "director": "The Wachowskis",
        "release_year": 1999,
        "rating": 8.7,
        "language": "English"
    }
)
```

```python
docs=[doc1,doc2,doc3,doc4,doc5]
```

```python
vector_store =Chrome(
    embedding_function=OpenAIEmbeddings(),
    persist_directory='chroma_db',
    collection_name="sample"
)
```

```python
# Add Documents
vector_store.add_documents(docs)
```

```python
#View Documents
vector_store.get(include=['embeddings','documents','metadata'])
```

```python
#Search Documents
vector_store.similarity_search(
    query='Movie with Thriller plot',
    k=2
)
```

```python
#Search with similarity score
vector_store.similarity_search_with_score(
    query='Horror movie',
    k=2
)
```

```python
#Meta-data Filtering
vector_store.similarity_seach_with_score(
    query="",
    filter={"genre":"Science Fiction"}
)
```

```python
#Update Documents
updated_doc1=Document(
    page_content="Jack dies trying to save Rose"
    metadata={"genre":"Neorealism"}
)
vector_store.update_document(document_id='write document id here',document=updated_doc1)
```
